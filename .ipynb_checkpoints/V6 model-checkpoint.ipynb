```python
import pandas as pd

# Load the dataset from Excel file
df = pd.read_excel('pvistats.xlsx')

# Handle missing values if any (for demonstration purposes)
df.fillna(method='ffill', inplace=True)  # Forward fill missing values

# Convert DateTime column to datetime object
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Explore the data
print(df.head())
print(df.info())
print(df.describe())
```

### Analyse Data 

```python
import matplotlib.pyplot as plt
data = df
# Reset index to ensure continuous datetime index
data.reset_index(drop=True, inplace=True)

list1 = ['DlyEn', 'TotEn', 'ParEn', 'WkE', 'MnE', 'YrE']

for i in list1:
    # Plotting the real data
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data[i], color='blue')
    plt.title('Energy Consumption Over Time')
    plt.xlabel('Date')
    plt.ylabel(i)
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame with the 'TotEn' column

# Calculate the first quartile (Q1)
Q1 = data['TotEn'].quantile(0.25)

# Calculate the third quartile (Q3)
Q3 = data['TotEn'].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define lower and upper bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the data points within the bounds
data_no_outliers = data[(data['TotEn'] >= lower_bound) & (data['TotEn'] <= upper_bound)]

# Plot the total energy consumption after outlier removal
plt.figure(figsize=(10, 6))
plt.plot(data_no_outliers['DateTime'], data_no_outliers['TotEn'], color='blue', label='Total Energy (No Outliers)')
plt.xlabel('DateTime')
plt.ylabel('Total Energy')
plt.title('Total Energy Consumption after Outlier Removal (IQR Method)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
count = 1
for i in data_no_outliers['TotEn']:
    if(i==0):
        count += 1
print(count)
```

```python
# Filter out zero values
data_no_outliers_no_zeros = data_no_outliers[data_no_outliers['TotEn'] != 0]

# Plot the total energy consumption after removing zero values
plt.figure(figsize=(10, 6))
plt.plot(data_no_outliers_no_zeros['DateTime'], data_no_outliers_no_zeros['TotEn'], color='blue', label='Total Energy (No Outliers or Zeros)')
plt.xlabel('DateTime')
plt.ylabel('Total Energy')
plt.title('Total Energy Consumption after Outlier Removal (IQR Method) and Zero Value Removal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming data_no_outliers_no_zeros is your DataFrame

# Convert DateTime column to datetime type if it's not already
data_no_outliers_no_zeros['DateTime'] = pd.to_datetime(data_no_outliers_no_zeros['DateTime'])

# Filter data
filtered_data = data_no_outliers_no_zeros[(data_no_outliers_no_zeros['DateTime'] > "2022-10-02 09:29:38") & (data_no_outliers_no_zeros['TotEn'] <= 2500.00)]

# Drop filtered rows
data_no_outliers_no_zeros = data_no_outliers_no_zeros.drop(filtered_data.index)

# Plot total energy
plt.figure(figsize=(10, 6))
plt.plot(data_no_outliers_no_zeros['DateTime'], data_no_outliers_no_zeros['TotEn'], color='blue')
plt.title('Total Energy Consumption over Time')
plt.xlabel('DateTime')
plt.ylabel('Total Energy Consumption')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Assuming data_no_outliers_no_zeros is your DataFrame

# Convert DateTime column to datetime type if it's not already
data_no_outliers_no_zeros['DateTime'] = pd.to_datetime(data_no_outliers_no_zeros['DateTime'])

# Set DateTime column as the index
data_no_outliers_no_zeros.set_index('DateTime', inplace=True)

# Resample data at monthly frequency and sum the energy consumption
monthly_energy_consumption = data_no_outliers_no_zeros['TotEn'].resample('M').sum()

# Print monthly energy consumption
print(monthly_energy_consumption)
```

```python
import matplotlib.pyplot as plt

# Plot monthly energy consumption
plt.figure(figsize=(10, 6))
monthly_energy_consumption.plot(kind='line', color='blue')
plt.title('Monthly Energy Consumption')
plt.xlabel('Month')
plt.ylabel('Energy Consumption')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filter out zero energy consumption values
monthly_energy_consumption_nonzero = monthly_energy_consumption[monthly_energy_consumption != 0]

# Plot monthly energy consumption without zero values
plt.figure(figsize=(10, 6))
monthly_energy_consumption_nonzero.plot(kind='line', color='blue')
plt.title('Monthly Energy Consumption (Non-zero values)')
plt.xlabel('Month')
plt.ylabel('Energy Consumption')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller test
result = adfuller(monthly_energy_consumption_nonzero)

# Extracting test statistic and p-value
adf_statistic = result[0]
p_value = result[1]

# Print the results
print("ADF Statistic:", adf_statistic)
print("p-value:", p_value)

# Check for significance
if p_value < 0.05:
    print("Reject the null hypothesis (H0). Data is stationary.")
else:
    print("Fail to reject the null hypothesis (H0). Data is non-stationary.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error

# Assuming you have monthly_energy_consumption_nonzero loaded
# If not, replace it with your actual data

# Define the training and testing data
train_data = monthly_energy_consumption_nonzero.iloc[:-12]  # Using data except the last 12 months for training
test_data = monthly_energy_consumption_nonzero.iloc[-12:]   # Using the last 12 months for testing
```

```python
# Fit SARIMA model
model = SARIMAX(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_model = model.fit(disp=False)
```

```python
# Predictions
predictions = sarima_model.predict(start=len(train_data), end=len(train_data)+len(test_data)-1, dynamic=False)
```

```python
# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(test_data, predictions)
print("Mean Squared Error (MSE):", mse)

# Visualize the actual vs. predicted values
plt.plot(train_data, label='Training Data')
plt.plot(test_data, label='Test Data')
plt.plot(predictions, label='Predictions')
plt.title('Monthly Energy Consumption Forecast using SARIMA')
plt.xlabel('Time')
plt.ylabel('Monthly Energy Consumption')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Ensure reproducibility
np.random.seed(42)

# Reshape data for LSTM model
X = np.array(monthly_energy_consumption_nonzero).reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler.fit_transform(X)

# Split data into train and test sets
train_size = int(len(X_scaled) * 0.8)
test_size = len(X_scaled) - train_size
train_data, test_data = X_scaled[0:train_size,:], X_scaled[train_size:len(X_scaled),:]

# Function to create dataset
def create_dataset(dataset, time_steps=1):
    X, Y = [], []
    for i in range(len(dataset)-time_steps-1):
        X.append(dataset[i:(i+time_steps), 0])
        Y.append(dataset[i + time_steps, 0])
    return np.array(X), np.array(Y)

# Create the dataset
time_steps = 1
X_train, Y_train = create_dataset(train_data, time_steps)
X_test, Y_test = create_dataset(test_data, time_steps)

# Reshape input to be [samples, time steps, features]
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

# Build LSTM model
model = Sequential()
model.add(LSTM(units=50, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, Y_train, epochs=100, batch_size=1, verbose=2)

# Make predictions
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

# Inverse scaling
train_predictions = scaler.inverse_transform(train_predictions)
test_predictions = scaler.inverse_transform(test_predictions)
Y_train = scaler.inverse_transform([Y_train])
Y_test = scaler.inverse_transform([Y_test])

# Calculate RMSE
train_rmse = np.sqrt(mean_squared_error(Y_train[0], train_predictions[:,0]))
test_rmse = np.sqrt(mean_squared_error(Y_test[0], test_predictions[:,0]))
print("Train RMSE:", train_rmse)
print("Test RMSE:", test_rmse)
# Function to create dataset with fixed length for X and Y
def create_dataset(dataset, time_steps=1):
    X, Y = [], []
    for i in range(len(dataset)-time_steps):
        X.append(dataset[i:(i+time_steps), 0])
        Y.append(dataset[i + time_steps, 0])
    return np.array(X), np.array(Y)


# Plot actual vs predicted
plt.figure(figsize=(10, 6))
plt.plot(monthly_energy_consumption_nonzero.index[1:train_size+1], Y_train[0], label='Actual (Train)')
plt.plot(monthly_energy_consumption_nonzero.index[train_size+2:], Y_test[0], label='Actual (Test)')
plt.plot(monthly_energy_consumption_nonzero.index[1:train_size+1], train_predictions[:,0], label='Predicted (Train)')
plt.plot(monthly_energy_consumption_nonzero.index[train_size+2:], test_predictions[:,0], label='Predicted (Test)')
plt.title('Actual vs Predicted Monthly Energy Consumption')
plt.xlabel('Date')
plt.ylabel('Energy Consumption')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Prepare the data
X = np.arange(len(monthly_energy_consumption_nonzero)).reshape(-1, 1)  # Index as input feature
y = monthly_energy_consumption_nonzero.values.reshape(-1, 1)

# Scale the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, shuffle=False)

# Build the model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=0)

# Make predictions
predictions_scaled = model.predict(X_test)
predictions = scaler.inverse_transform(predictions_scaled)

# Plot actual vs predicted
plt.figure(figsize=(10, 6))
plt.plot(monthly_energy_consumption_nonzero.index[-len(predictions):], y_test.reshape(-1), label='Actual')
plt.plot(monthly_energy_consumption_nonzero.index[-len(predictions):], predictions.reshape(-1), label='Predicted')
plt.title('Actual vs Predicted Monthly Energy Consumption')
plt.xlabel('Date')
plt.ylabel('Energy Consumption')
plt.legend()
plt.show()

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, predictions_scaled))
print("Root Mean Squared Error (RMSE):", rmse)
```
