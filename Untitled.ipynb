```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
# Load the dataset
data = pd.read_excel('pvistats.xlsx')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['DateTime'], data['DlyEn'], marker='o', linestyle='-')
plt.title('Daily Energy Consumption')
plt.xlabel('Date')
plt.ylabel('Energy Consumption')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['DateTime'], data['TotEn'], marker='o', linestyle='-')
plt.title('Total Energy Consumption')
plt.xlabel('Date')
plt.ylabel('Energy Consumption')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.plot(data['DateTime'], data['ParEn'], marker='o', linestyle='-')
plt.title('Energy Consumption per parameter')
plt.xlabel('Date')
plt.ylabel('Energy Consumption')
plt.grid(True)
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
data_hourly = data.resample('H', on='DateTime').mean()
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(data_hourly.index, data_hourly['DlyEn'], marker='o', linestyle='-')
plt.title('Hourly Average Energy Consumption')
plt.xlabel('Date and Time')
plt.ylabel('Energy Consumption')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
