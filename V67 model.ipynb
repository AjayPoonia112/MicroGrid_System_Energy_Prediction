```python
import pandas as pd

# Load the dataset from Excel file
df = pd.read_excel('pvistats.xlsx')

# Handle missing values if any (for demonstration purposes)
df.fillna(method='ffill', inplace=True)  # Forward fill missing values

# Convert DateTime column to datetime object
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Explore the data
print(df.head())
print(df.info())
print(df.describe())
```

### Analyse Data 

```python
import matplotlib.pyplot as plt
data = df
# Reset index to ensure continuous datetime index
data.reset_index(drop=True, inplace=True)

list1 = ['DlyEn', 'TotEn', 'ParEn', 'WkE', 'MnE', 'YrE']

for i in list1:
    # Plotting the real data
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data[i], color='blue')
    plt.title('Energy Consumption Over Time')
    plt.xlabel('Date')
    plt.ylabel(i)
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame with the 'TotEn' column

# Calculate the first quartile (Q1)
Q1 = data['TotEn'].quantile(0.25)

# Calculate the third quartile (Q3)
Q3 = data['TotEn'].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define lower and upper bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the data points within the bounds
data_no_outliers = data[(data['TotEn'] >= lower_bound) & (data['TotEn'] <= upper_bound)]

# Plot the total energy consumption after outlier removal
plt.figure(figsize=(10, 6))
plt.plot(data_no_outliers['DateTime'], data_no_outliers['TotEn'], color='blue', label='Total Energy (No Outliers)')
plt.xlabel('DateTime')
plt.ylabel('Total Energy')
plt.title('Total Energy Consumption after Outlier Removal (IQR Method)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filter out zero values
data_no_outliers_no_zeros = data_no_outliers[data_no_outliers['TotEn'] != 0]

# Plot the total energy consumption after removing zero values
plt.figure(figsize=(10, 6))
plt.plot(data_no_outliers_no_zeros['DateTime'], data_no_outliers_no_zeros['TotEn'], color='blue', label='Total Energy (No Outliers or Zeros)')
plt.xlabel('DateTime')
plt.ylabel('Total Energy')
plt.title('Total Energy Consumption after Outlier Removal (IQR Method) and Zero Value Removal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Assuming data_no_outliers_no_zeros is your DataFrame

# Convert DateTime column to datetime type if it's not already
data_no_outliers_no_zeros['DateTime'] = pd.to_datetime(data_no_outliers_no_zeros['DateTime'])

# Set DateTime column as the index
data_no_outliers_no_zeros.set_index('DateTime', inplace=True)

# Resample data at monthly frequency and sum the energy consumption
monthly_energy_consumption = data_no_outliers_no_zeros['TotEn'].resample('M').sum()

# Print monthly energy consumption
print(monthly_energy_consumption)
```

```python
import matplotlib.pyplot as plt

# Plot monthly energy consumption
plt.figure(figsize=(10, 6))
monthly_energy_consumption.plot(kind='line', color='blue')
plt.title('Monthly Energy Consumption')
plt.xlabel('Month')
plt.ylabel('Energy Consumption')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
# Replace zero energy consumption values with NaN
monthly_energy_consumption_nonzero = monthly_energy_consumption.replace(0, np.nan)

# Plot monthly energy consumption with NaN instead of zero values
plt.figure(figsize=(10, 6))
monthly_energy_consumption_nonzero.plot(kind='line', color='blue')
plt.title('Monthly Energy Consumption (Non-zero values)')
plt.xlabel('Month')
plt.ylabel('Energy Consumption')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Replace NaN values with the mean of non-NaN values
monthly_energy_consumption_nonzero_filled = monthly_energy_consumption_nonzero.fillna(monthly_energy_consumption_nonzero.mean())

# Plot monthly energy consumption with NaN replaced by mean
plt.figure(figsize=(10, 6))
monthly_energy_consumption_nonzero_filled.plot(kind='line', color='blue')
plt.title('Monthly Energy Consumption (Non-zero values)')
plt.xlabel('Month')
plt.ylabel('Energy Consumption')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Normalize the data
monthly_energy_consumption_normalized = (monthly_energy_consumption_nonzero_filled - monthly_energy_consumption_nonzero_filled.mean()) / monthly_energy_consumption_nonzero_filled.std()

# Plot normalized monthly energy consumption
plt.figure(figsize=(10, 6))
monthly_energy_consumption_normalized.plot(kind='line', color='blue')
plt.title('Normalized Monthly Energy Consumption')
plt.xlabel('Month')
plt.ylabel('Normalized Energy Consumption')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def test_stationary(timeseries):
    rolmean = timeseries.rolling(window=6).mean()
    rolstd = timeseries.rolling(window=6).std()

    orig = plt.plot(timeseries, label="Original")
    mean = plt.plot(rolmean, label='Rolling Mean')
    std = plt.plot(rolstd, label='Rolling std')
    
    from statsmodels.tsa.stattools import adfuller
    plt.legend(loc='best')
    plt.show()
    
    dftest = adfuller(timeseries)
    print(dftest)
```

```python
# Apply logarithmic transformation to the filled data
monthly_energy_consumption_log = np.log(monthly_energy_consumption_nonzero_filled)

# Plot monthly energy consumption with logarithmic scale
plt.figure(figsize=(10, 6))
monthly_energy_consumption_log.plot(kind='line', color='blue')
plt.title('Monthly Energy Consumption (Non-zero values) - Log Scaled')
plt.xlabel('Month')
plt.ylabel('Log Energy Consumption')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate rolling mean and standard deviation
rolling_mean = monthly_energy_consumption_log.rolling(window=3).mean()
rolling_std = monthly_energy_consumption_log.rolling(window=3).std()

# Plot log-scaled monthly energy consumption with rolling mean and standard deviation
plt.figure(figsize=(10, 6))
plt.plot(monthly_energy_consumption_log, label='Log Energy Consumption')
plt.plot(rolling_mean, label='Rolling Mean', color='red')
plt.plot(rolling_std, label='Rolling Std', color='green')
plt.title('Log Scaled Monthly Energy Consumption with Rolling Mean and Std')
plt.xlabel('Month')
plt.ylabel('Log Energy Consumption')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
ma = monthly_energy_consumption_log.rolling(window=6).std()
monthly_energy_consumption_log_minus_ma = monthly_energy_consumption_log - ma
```

```python
# Plot the filled data
plt.figure(figsize=(10, 6))
monthly_energy_consumption_log_minus_ma_no_NA.plot(kind='line')
plt.title('Monthly Energy Consumption (Log Minus Moving Average) - Filled')
plt.xlabel('Date')
plt.ylabel('Log Energy Consumption')
plt.legend(loc='upper left')
plt.show()
```

```python
# Calculate rolling mean and standard deviation
rolling_mean = monthly_energy_consumption_log_minus_ma_no_NA.rolling(window=3).mean()
rolling_std = monthly_energy_consumption_log_minus_ma_no_NA.rolling(window=3).std()

# Plot log-scaled monthly energy consumption with rolling mean and standard deviation
plt.figure(figsize=(10, 6))
plt.plot(monthly_energy_consumption_log_minus_ma_no_NA, label='Log Energy Consumption')
plt.plot(rolling_mean, label='Rolling Mean', color='red')
plt.plot(rolling_std, label='Rolling Std', color='green')
plt.title('Log Scaled Monthly Energy Consumption with Rolling Mean and Std')
plt.xlabel('Month')
plt.ylabel('Log Energy Consumption')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller test
result = adfuller(monthly_energy_consumption_log_minus_ma_no_NA)

# Extract test statistics and p-value
adf_statistic = result[0]
p_value = result[1]

# Print the results
print('ADF Statistic:', adf_statistic)
print('p-value:', p_value)

# Interpret the results
if p_value < 0.05:
    print('Reject the null hypothesis. The data is stationary.')
else:
    print('Fail to reject the null hypothesis. The data is non-stationary.')
```
